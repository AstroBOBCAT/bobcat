# Log-Log Plots of Sensitivity

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## PPTA

```python
dr1 = pd.read_csv('PPTAdr1_hsensitivity.txt')
dr2 = pd.read_csv('PPTAdr2_hsensitivity.txt')
```

```python
dr1.head()
```

```python
#plotting

x1 = dr1['x']
y1 = dr1['y']
plt.loglog(x1,y1,label='Data Release 1')

x2 = dr2['x']
y2 = dr2['y']
plt.loglog(x2,y2,label='Data Release 2')

plt.title('PPTA Sensitivity Curves')
plt.xlabel('Gravitational Wave Frequency (Hz)')
plt.ylabel('Strain Amplitude')
plt.legend()
```

## NANOgrav

```python
ng5 = pd.read_csv('5yr_ul_skyavg.txt')
ng9 = pd.read_csv('9yr_ul_skyavg.txt')
ng11 = pd.read_csv('11yr_ul_skyavg.txt')
ng11_best_sky = pd.read_csv('11yr_ul_bestskyloc.txt')
```

```python
ng5.head()
```

```python
#plotting
plt.figure(figsize=(8,6))
x1 = ng5['GW Freq']
y1 = ng5['Strain UL']
plt.loglog(x1,y1,label='NANOgrav 5yr')

x2 = ng9['GW Freq']
y2 = ng9['Strain UL']
plt.loglog(x2,y2,label='NANOgrav 9yr')

x3 = ng11['GW Freq']
y3 = ng11['Strain UL']
plt.loglog(x3,y3,label='NANOgrav 11yr')

x4 = ng11_best_sky['GW Freq']
y4 = ng11_best_sky['Strain UL']
plt.loglog(x4,y4,label='NANOgrav 11yr Best Sky Location')

plt.title('NANOgrav Sensitivity Curves')
plt.xlabel('Gravitational Wave Frequency (Hz)')
plt.ylabel('Strain Amplitude')
plt.legend()
```

## Combined

```python
#plotting combined
plt.figure(figsize=(9,7))
#PPTA
p1 = dr1['x']
q1 = dr1['y']
plt.loglog(p1,q1,label='PPTA Data Release 1')
p2 = dr2['x']
q2 = dr2['y']
plt.loglog(p2,q2,label='PPTA Data Release 2')

#NANOgrav
x1 = ng5['GW Freq']
y1 = ng5['Strain UL']
plt.loglog(x1,y1,label='NANOgrav 5yr')
x2 = ng9['GW Freq']
y2 = ng9['Strain UL']
plt.loglog(x2,y2,label='NANOgrav 9yr')
x3 = ng11['GW Freq']
y3 = ng11['Strain UL']
plt.loglog(x3,y3,label='NANOgrav 11yr')
x4 = ng11_best_sky['GW Freq']
y4 = ng11_best_sky['Strain UL']
plt.loglog(x4,y4,label='NANOgrav 11yr Best Sky Location')

#format
plt.title('NANOgrav Sensitivity Curves')
plt.xlabel('Gravitational Wave Frequency (Hz)')
plt.ylabel('Strain Amplitude')
plt.legend()
```
